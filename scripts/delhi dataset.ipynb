```python
import pandas as pd
```

```python
df=pd.read_csv("delhi.csv")
df.head()
```

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
df=df.rename(columns={"1":"Borough","2":"Neighborhood"},errors="raise")
```

```python
df.to_csv("Delhi.csv")
```

```python
df1=df[df.Borough == "DWARKA"]
```

```python
df2=df[df.Borough == "NEW DELHI"]
```

```python
delh=[df1,df2]
```

```python
delhi=pd.concat(delh)
```

```python
delhi=delhi.reset_index(drop=True)
delhi.shape
```

```python
lat=[]
lon=[]
```

```python
for x in delhi.Neighborhood:
    location=None
    
    while location is None:
        geolocator=Nominatim(user_agent="shivam")
        location=geolocator.geocode('{}, Delhi, India'.format(x))
        x=str(x).split(" ")[:-1]
    lat.append(location.latitude)
    print(location)
    print(location.latitude)
    lon.append(location.longitude)
    print(location.longitude)
```
